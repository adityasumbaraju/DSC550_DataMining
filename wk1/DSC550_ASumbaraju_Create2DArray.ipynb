1. Create an array of size [10,50] and use the random function to fill it with integers between 1 and 500 (inclusive). Load it to a DataFrame.
 
2. Calculate the sum of each row; the sum of each column; and the sum of all entries.

3. Determine the minimum; maximum; average; and standard deviation of all entries.

4. Sort the DataFrame: 
   a) sort on column 2
   b) sort on row 2

# ---------------------------------------------------------------------------------------------------------


```python
# import libraries
import numpy as np
import pandas as pd

# 1
# Create an array with random integers
# try for smaller arrays to check the pointers in assignment
# rows, cols = (3,3) # took 3 rows & 3 columns to match the requirement 4 of assignment for testing.
# switching to  size [10,50] as the intial testing was successful with smaller set.
rows, cols = (10,50)
rint_arr = [[np.random.randint(1,500) for i in range(cols)] for j in range(rows)]
# Load into DataFrame
df = pd.DataFrame(rint_arr)
print("Randomized Integers Data Frame")
df

```

```python
#  Calculate the sum of each row; the sum of each column; and the sum of all entries.

df.loc['Total',:] = df.sum(axis=0)
df.loc[:,'Total'] = df.sum(axis=1)
print("Sum of each Row, Column, and all Values")
df

```

```python
# 3 Determine the minimum; maximum; average; and standard deviation of all entries.
# Minimum of all random integers from dataframe
print("\nMinimum Value: " + str(df.values.min())) #min() will provide the minimum value

# Maximum of all entries
print("Maximum Value: " + str(df.values.max()))  #max() will provide the maximum value

# Average of all entries
print("Average Value: " + str(df.values.mean()))  #mean() will provide the mean value

# Standard Deviation of all entries
print("Standard Deviation: " + str(df.values.std()) + "\n") #std() will provide the Standard Deviation 

```

```python
# Sort the DataFrame: a) sort on column 2 b) sort on row 2
# Sort dataframe on row_index 2
print("Sort by Row index 2")
df_row_sort=df.sort_values(by=2, axis=1)
df_row_sort
```

```python
# Sort dataframe on column_index 2
print("Sort by Column index 2")
df_col_sort=df.sort_values(by=2, axis=0)
df_col_sort
```
