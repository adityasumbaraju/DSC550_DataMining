## Part 1

```python

# Import libraries
import pandas as pd
import numpy as np
import string
import nltk

from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from nltk.tokenize import word_tokenize

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.preprocessing import MultiLabelBinarizer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# Downloaded the file from Week 2 Data Files andsaved in my local at C:\BU\DSC550\wk2\controversial-comments.jsonl
# # Pre-processing Text
def readFile(f='C:\BU\DSC550\wk2\controversial-comments.jsonl'):
    df = pd.read_json(f, lines=True)
    return df
    
df_controversial_comments = readFile()
df_controversial_comments
```

##  Convert text to lowercase.

```python
df_controversial_comments['txt'] = df_controversial_comments['txt'].str.lower() # function lower() converts the text to lower case
df_controversial_comments
```

## Remove all punctuation from the text.

```python
# strip off punctuation characters as shown below
# '!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~'
df_controversial_comments['txt'] = df_controversial_comments['txt'].str.translate(str.maketrans('','',string.punctuation))
df_controversial_comments
```

## Remove stop words

```python
# Exclude stopwords using nltk library with Python's list comprehension and pandas.DataFrame.apply.
# using lambda function tha will split, remove any stopwords and then stitches back the original txt
stop_words=stopwords.words('english')

df_controversial_comments['txt'] = df_controversial_comments['txt'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop_words)]))
df_controversial_comments
```

## Apply NLTK’s PorterStemmer

```python
porter = PorterStemmer()
df_controversial_comments['txt'] = df_controversial_comments['txt'].apply(lambda x: ' '.join([porter.stem(word) for word in x.split()]))
df_controversial_comments
```

```python
# download data to local for final verification
df_controversial_comments.to_csv("C:\BU\DSC550\wk2\clean_controversial_comments.csv")
```

##  Convert each text entry into a word-count vector 

```python
# Create the bag of words feature matrix
count = CountVectorizer()
# Reducing the size of data to execute the task
# Reason:
# MemoryError: Unable to allocate 1.56 TiB for an array with shape (950000, 226148) and data type int64
df_sample_controversial_comments = df_controversial_comments.sample(frac = 0.005) # analyzing with 0.5 % of data from cleaned dataframe
df_sample_controversial_comments
```

## Convert each text entry into a part-of-speech tag vector 

```python
words = df_sample_controversial_comments["txt"].values.flatten()
# Array declaration
part_of_speechTag = []

# Word tagging
for word in words:
    word_tag = nltk.pos_tag(word_tokenize(word))
    part_of_speechTag.append([tag for word, tag in word_tag])

# Feature Conversion:use one hot encoding to convert the tags into features
one_hot_multi = MultiLabelBinarizer()
one_hot_multi.fit_transform(part_of_speechTag)
```

## Convert each entry into a term frequency-inverse document frequency (tfidf) vector 
#reference: (see section 6.9 in the Machine Learning with Python Cookbook).

```python
tfidf = TfidfVectorizer()
# Term Frequ-inverse document vector
feature_matrix = tfidf.fit_transform(df_controversial_comments["txt"])
feature_matrix
```
